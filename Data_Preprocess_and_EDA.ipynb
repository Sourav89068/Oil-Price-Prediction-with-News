# Analysis of Covid-situation

```python
def change(news,percentage):
    date = []
    for i in range(len(((np.abs((news.Close-news.Close.shift(1)))/news.Close)[1:]*100).values))  :
        if ((np.abs((news.Close-news.Close.shift(1)))/news.Close)[1:]*100).values[i]>percentage:
            date.append((news.iloc[i,2]))
    return date
```

```python
covid_situation = data[data['date'].isin(change(wti_data,5))].tail(37)
```
